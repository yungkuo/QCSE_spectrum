```python
% matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import libtiff
from sub import lambda_cali_v1
import os
```

```python
filePath = '/Users/yungkuo/Documents/Data/ND/092415/'
```

```python
scan_w = 6      
displacement = scan_w*2
x = np.arange(0,512,1)
```

```python
c1 = filePath+'c1.tif'
mov = libtiff.TiffFile(c1)
c1 = mov.get_tiff_array()
c1 = np.mean(c1[0,50:,:],dtype='d', axis=0)-np.mean(c1[0,0:20,:],dtype='d', axis=0)

c2 = filePath+'c2.tif'
mov = libtiff.TiffFile(c2)
c2 = mov.get_tiff_array()
c2 = np.mean(c2[0,50:,:],dtype='d', axis=0)-np.mean(c2[0,0:20,:],dtype='d', axis=0)

c3 = filePath+'c3.tif'
mov = libtiff.TiffFile(c3)
c3 = mov.get_tiff_array()
c3 = np.mean(c3[0,50:,:],dtype='d', axis=0)-np.mean(c3[0,0:20,:],dtype='d', axis=0)

lamp = filePath+'lamp.tif'
mov = libtiff.TiffFile(lamp)
lamp = mov.get_tiff_array()
lamp = np.mean(lamp[0,20:,:],dtype='d', axis=0)-np.mean(lamp[0,0:20,:],dtype='d', axis=0)

x_lambda = lambda_cali_v1.x_lambda(lamp, c1, c2, c3, x)
```

```python
pts = np.array([[ 329.,  316.],
                [ 325.,  459.],
                [ 329.,  427.]])
```

```python
def listdir_nohidden(path):
    for f in os.listdir(path):
        if f.endswith('4Hz.tif'):
            yield f
```

```python
for n in range(len(pts)):
    fig, ax = plt.subplots(2,4,figsize=(4,8))
    for count, file in enumerate(listdir_nohidden(filePath)):
        current_file = os.path.join(filePath, file)
        mov = libtiff.TiffFile(current_file)
        movie = mov.get_tiff_array()
        movie = np.array(movie[:,:,:],dtype='d')
        if count == 0:
            frame = len(movie[:,0,0])
            mean_I = np.mean(movie, axis=0)
            ax[0,0].imshow(mean_I, cmap='afmhot', interpolation='None') 
            ax[0,0].axhline(y=pts[n,1]-scan_w, xmin=0, xmax=512, alpha=0.5, c= 'w')
            ax[0,0].axhline(y=pts[n,1]+scan_w, xmin=0, xmax=512, alpha=0.5, c= 'w')
            ax[0,0].annotate(n, xy=(pts[n,0], pts[n,1]), xytext=(pts[n,0], pts[n,1]+20),color='w')  
            ax[0,0].set_title('Mean image')
        V = file.split('V')[0]
        spectra = np.mean(movie[:,pts[n,1]-scan_w:pts[n,1]+scan_w,:] , axis=1)
        bgspectra = np.mean(movie[:,pts[n,1]-scan_w-displacement:pts[n,1]+scan_w-displacement,:] , axis=1)
        spectra_bgcr = spectra - bgspectra    
        Von_spec = np.mean([spectra_bgcr[i,:] for i in range(frame) if i%2==0],  axis=0)
        Voff_spec = np.mean([spectra_bgcr[i,:] for i in range(frame) if i%2==0], axis=0)
        ax[(count+1)/4,(count+1)%4].plot(x_lambda, Von_spec, '-r', label='Von')
        ax[(count+1)/4,(count+1)%4].plot(x_lambda, Voff_spec,  '--b', label='Voff')
        ax[(count+1)/4,(count+1)%4].set_title('{}V'.format(V))
        ax[(count+1)/4,(count+1)%4].legend(bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
        fig.canvas.draw()
```
